```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
amp = 1; sr = 10000; phase = 0

Eb4 = 311.13
G4 = 392.00
Bb4 = 466.16
B4 = 493.88
Eb5 = 622.25
E5 = 659.25
F5 = 698.46
G5 = 783.99
Bb5 = 932.33
B5 = 987.77
C6 = 1046.50
Eb6 = 1244.51
```

```python
# MapleStory Lith Harbor OST
# 8분음표 dur = 0.25 기준

notes = [B4, F5, Eb6, C6, Eb6, B5, F5, \
         B4, F5, Eb6, C6, Eb6, B5, F5, \
         B4, E5, F5, B5, F5, E5, F5, \
         B4, E5, F5, B5, F5, E5, F5, \

         Eb4, Bb4, E5, F5, G5, Bb5, 0, \
         E5, F5, 0, G4, Bb4, Eb5, E5, \
         G5, 0, F5, E5, Bb5]

durs = [0.25, 0.25, 0.25, 0.5, 0.25, 0.25, 0.25, \
        0.25, 0.25, 0.25, 0.5, 0.25, 0.25, 0.25, \
        0.25, 0.25, 0.25, 0.5, 0.25, 0.25, 0.25, \
        0.25, 0.25, 0.25, 0.5, 0.25, 0.25, 0.25, \

        0.25, 0.25, 0.25, 0.75, 0.25, 0.75, 0.75, \
        0.5, 0.25, 0.25, 0.25, 0.25, 0.75, 0.25, \
        0.75, 0.25, 0.5, 0.5, 0.25]
```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
  s, t, th= phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all, s))
```

```python
ipd.Audio(s_all, rate=sr)
```
